Importing libraries and dataset

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("adult.data", skipinitialspace = True, header = None)
```

```python
data.head()
```

The data needs to be cleaned and formatted.

1.) Adding column names to the dataframe.

```python
columns = ["age", "workclass", "fnlwgt", "education", "education-num", "marital-status",
           "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss",
           "hours-per-week", "native-country", "income"]
data.columns = columns
data.head()
```

2.) The missing values in the dataset are stored as '?'. Hence, cleaning the data by replacing ? by NaN and then dropping the null values.

```python
data = data.replace("?", np.NaN)
data.head()
```

```python
# Checking missing values
data.isnull().sum()
```

```python
print("Shape of data before dropping records with NA values : ", data.shape)
```

```python
data.dropna(axis = 0, inplace = True)
data = data.reset_index(drop = True)
data.head()
```

```python
print("Shape of data before dropping records with NA values : ", data.shape)
```

```python
data.isnull().sum() # missing values dropped
```
